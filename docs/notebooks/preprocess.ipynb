```python
import novaice
import anndata as ad
import pandas as pd
import numpy as np
```

## Create anndata object

```python
obs = pd.DataFrame(
    {
        "smiles": [
            "CC(=O)OC1=CC=CC=C1C(=O)O",
            "CC(C)CC1=CC=C(C=C1)C(C)C(=O)O",
            "C([C@@H]1[C@H]([C@@H]([C@H](C(O1)O)O)O)O)O",
        ],
        "molecule_name": ["aspirin", "ibuprofene", "glucose"],
    }
)
obs
```

```python
adata = ad.AnnData(X=np.zeros(shape=(3, 5)), obs=obs)
```

```python
novaice.pp.foundation_model_embeddings(
    adata, smiles_col="smiles", key_added="molformer", model_name="ibm-research/MoLFormer-XL-both-10pct", device="cpu"
)
adata.obsm["molformer"].shape
```

```python
novaice.pp.morgan_fingerprints(adata, smiles_col="smiles", key_added="fingerprint", radius=2, n_bits=4096)
adata.obsm["fingerprint"].shape
```
